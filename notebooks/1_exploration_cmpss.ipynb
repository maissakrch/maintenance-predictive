```python
import pandas as pd
import matplotlib.pyplot as plt

# Définir les colonnes
cols = ['unit', 'cycle', 'op_setting_1', 'op_setting_2', 'op_setting_3'] + \
       [f'sensor_{i}' for i in range(1, 22)]

# Charger le fichier
df = pd.read_csv('../data/raw/train_FD001.txt', sep='\s+', header=None)
df.columns = cols

# Aperçu des données
print("Aperçu des données :")
display(df.head())

# Nombre de moteurs
nb_moteurs = df['unit'].nunique()
print(f"Nombre de moteurs différents : {nb_moteurs}")

# Distribution du nombre de cycles avant panne
cycles_par_moteur = df.groupby('unit')['cycle'].max()
print("Durée de vie moyenne :", round(cycles_par_moteur.mean(), 2))

# Affichage graphique
cycles_par_moteur.hist(bins=30)
plt.title("Distribution du nombre de cycles avant panne")
plt.xlabel("Nombre de cycles")
plt.ylabel("Nombre de moteurs")
plt.grid(True)
plt.show()
```
